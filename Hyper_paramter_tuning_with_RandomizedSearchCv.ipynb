```
import keras
import sklearn
from keras import layers,models,optimizers,losses,activations,regularizers,initializers
from keras.utils import to_categorical
import pandas as pd
import numpy as np
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import LabelEncoder        # for label encoder
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import GridSearchCV
```

```
seed=7
np.random.seed(seed)
```

```
iris = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data',
                   names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'label'])
```

```
le = LabelEncoder()
iris["label"]=le.fit_transform(iris["label"])
```

```
x=np.array(iris.drop(["label"],axis=1))
y=np.array(iris["label"])
```

```
x=x.astype("float32")
y=to_categorical(y, num_classes=3)
```

```
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
iris.head() 
```

```
def base_model():
  model=models.Sequential()
  model.add(layers.Dense(32,activation="relu",kernel_initializer='glorot_uniform'))
  model.add(layers.Dense(16,activation="relu",kernel_initializer='glorot_uniform'))
  model.add(layers.Dense(3,activation="softmax"))
  opt = keras.optimizers.Adam(lr=0.001, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=False)
  model.compile(loss="categorical_crossentropy",optimizer=opt,metrics=["accuracy"])
  return model
```

```
model=KerasClassifier(build_fn=base_model,epochs=50,batch_size=32)
history=model.fit(X_train,y_train)
```

```
# RandomizedSearchCV
batch_sizes = [16,32,64,128,512]
epochs = [5, 10, 50]
params = {'batch_size': batch_sizes, 'epochs': epochs}
model=KerasClassifier(build_fn=base_model)
clf = GridSearchCV(model, params)
#clf=RandomizedSearchCV(model,params,cv=10,scoring="accuracy",n_iter=10,random_state=5)
clf.fit(X_train,y_train)
```

```
print(clf.best_score_, clf.best_params_,"\n","\n")
means = clf.cv_results_['mean_test_score']
stds = clf.cv_results_['std_test_score']
parameters = clf.cv_results_['params']
for mean, stds, parameters in zip(means, stds, parameters):
    print("%f (%f) with: %r" % (mean,stds, parameters))
```

```
# RandomizedSearchCV
batch_sizes = [16,32,64,128,512]
epochs = [5, 10, 50]
params = {'batch_size': batch_sizes, 'epochs': epochs}
model=KerasClassifier(build_fn=base_model)
clf=RandomizedSearchCV(model,params)
clf.fit(X_train,y_train)
pd.DataFrame(clf.cv_results_)[['mean_test_score', 'std_test_score', 'params']]
```

```
#tuning Learning rate optimizer ,neuron
def base_model2(neuron=16,lRate=0.001,WIntializer="gloroto"):
    model=models.Sequential()
    model.add(layers.Dense(neuron,activation="relu",kernel_initializer=WIntializer))
    model.add(layers.Dense(neuron,activation="relu",kernel_initializer=WIntializer))
    model.add(layers.Dense(3,activation="softmax"))
    opt = keras.optimizers.Adam(lr=lRate, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=False)
    model.compile(loss="categorical_crossentropy",optimizer=opt,metrics=["accuracy"])
    return model
```

```
# RandomizedSearchCV
batch_sizes = [16,32,64,128,512]
epochs = [5, 10, 50]
neuron = [1, 32, 16, 8, 4, 2]
WIntializer = ['uniform', 'lecun_uniform', 'normal', 'zero', 'glorot_normal', 'glorot_uniform', 'he_normal', 'he_uniform']
lRate = [0.0001,0.001,0.01,0.1,1]
params = {'batch_size': batch_sizes,
          'epochs': epochs,
          "WIntializer" : WIntializer,
          "neuron" : neuron,
          "lRate":lRate
         }
model=KerasClassifier(build_fn=base_model2)
clf=RandomizedSearchCV(model,params)
clf.fit(X_train,y_train)
```

```
print(pd.DataFrame(clf.cv_results_)[['mean_test_score', 'std_test_score', 'params']])
print(clf.best_score_)
print(clf.best_params_)
```

```
#tuning Learning rate optimizer ,neuron
def base_model3(neuron,lRate):
    model=models.Sequential()
    model.add(layers.Dense(neuron,activation="relu",kernel_initializer='glorot_uniform'))
    model.add(layers.Dense(neuron,activation="relu",kernel_initializer='glorot_uniform'))
    model.add(layers.Dense(3,activation="softmax"))
    opt = keras.optimizers.Adam(lr=lRate, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=False)
    model.compile(loss="categorical_crossentropy",optimizer=opt,metrics=["accuracy"])
    return model
```

```

# RandomizedSearchCV
neuron = [1, 32, 16, 8, 4, 2]
lRate = [0.0001,0.001,0.01,0.1,1.0]
params = {
          "neuron" : neuron,
          "lRate":lRate
         }
model=KerasClassifier(build_fn=base_model3,epochs=50,batch_size=32)
clf=RandomizedSearchCV(model,params)
clf.fit(X_train,y_train)
```

```
print(clf.best_score_,clf.best_params_)
pd.DataFrame(clf.cv_results_)[['mean_test_score', 'std_test_score', 'params']]
```
